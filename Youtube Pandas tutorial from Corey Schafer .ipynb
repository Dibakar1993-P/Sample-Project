```python
country_grp['LanguageWorkedWith'].apply( lambda x: x.str.contains('Python').sum())
```

```python
country_respondents= df['Country'].value_counts()
country_uses_python = country_grp['LanguageWorkedWith'].apply( lambda x: x.str.contains('Python').sum())
python_df=pd.concat([country_respondents,country_uses_python],axis='columns',sort=False)
python_df.rename(columns={'Country':'Number of Respondents','LanguageWorkedWith':'Number of people Knows Python'},inplace=True)
python_df['Percentage of people knows Python']=(python_df['Number of people Knows Python']/python_df['Number of Respondents'])*100
python_df.sort_values(by='Percentage of people knows Python',ascending=True,inplace=True)
python_df.loc['Bangladesh']
```
